```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("diabetes_full.csv")
df.head()

x=df.iloc[:,:5]
y=df.iloc[:,5:]

print(x)
print(y)

x.head(df.size)

y.head(df.size)

from sklearn import tree
model=tree.DecisionTreeClassifier(criterion="entropy")
model.fit(x.values,y)
```

```python
model.score(x.values,y)
```

```python
model.predict([[150,66,342,34.7,42]])
```

```python
plt.figure(figsize=(18,18))
tree.plot_tree(model, fontsize=5)
plt.show()
```
